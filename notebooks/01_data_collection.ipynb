# Phase 1: Data Collection

Objective:
To collect 5 years of historical stock data for selected NSE-listed companies 
using the yfinance API.

Dataset includes:
- Open
- High
- Low
- Close
- Adjusted Close
- Volume

This dataset will be used later for:
- Feature engineering
- Trend classification
- Risk modelling
- Machine learning model training

```python
import yfinance as yf
import pandas as pd
```

```python
# NSE-listed large cap stocks (sector diversity)
tickers = ["RELIANCE.NS", "TCS.NS", "HDFCBANK.NS"]

tickers
```

```python
for ticker in tickers:
    print(f"Downloading {ticker}...")
    df = yf.download(ticker, period="5y")
    df.to_csv(f"../data/raw/{ticker}.csv")
    
print("Download Complete ✅")
```

```python
df = pd.read_csv("../data/raw/RELIANCE.NS.csv")

df.head()
```
